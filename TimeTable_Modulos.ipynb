```python
from PIL import Image
from PIL import ImageDraw
from math import sin, cos, sqrt, radians
from IPython.display import display
```

```python
size = (800, 800)
radius = sqrt((size[0] ** 2) + (size[1] ** 2)) // 3.5
count = 553
k = 7
print(radius)
```

```python
theta = 360 / count
```

```python
def circle(image, center, radius, width = 1):
    draw = ImageDraw.Draw(im = image)
    leftCorner = (center[0] - radius, center[1] - radius)
    rightCorner = (center[0] + radius, center[1] + radius)

    boundingBox = [leftCorner, rightCorner]
    draw.ellipse(xy = boundingBox, fill = None, outline = (255, 255, 255), width = width)
```

```python
def line(image, start, end, width = 1):
    draw = ImageDraw.Draw(im = image)
    draw.line(xy = [start, end], fill = (0, 0, 255), width = width)
```

```python
# Creating the coodinate list
coordinates = []
shiftX = -(size[0] / 2)
shiftY = size[1] / 2

for i in range(count):
    rad = radians(i * theta)
    coordinates.append(((radius * cos(rad)) - shiftX, shiftY - (radius * sin(rad))))
```

```python
# print(coordinates)
```

```python
baseImg = Image.new(mode = 'RGB', size = size)
display(baseImg)
```

```python
circle(image = baseImg, center = (size[0] // 2, size[1] // 2), radius = radius, width = 2)
display(baseImg)
```

```python
for i in range(count):
    line(image = baseImg, start = coordinates[i], end = coordinates[(i * k) % count])
```
